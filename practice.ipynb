````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from random import *
import random
import matplotlib.font_manager as fm
from numpy.linalg import inv, qr
import json
import sys
import requests
from collections import defaultdict, Counter
import seaborn as sns
from numpy.random import randn
````

````python
fig = plt.figure()                     # 주피터는 한셀에 모두 입력해야 출력됨
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

plt.plot(np.random.randn(50).cumsum(), 'b--')   # 검정색 점선 그래프
_ = ax1.hist(np.random.randn(100), bins=20, color='g', alpha=0.3)     # 검정색 히스토그램
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30), color='r')   # 산점도
````

````python
fig, axes = plt.subplots(2,3)
````

````python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500), bins=50, color='g', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
````

````python
plt.plot(randn(30).cumsum(), 'ro--')  # 선그래프 마커, 점찍기, o
````

````python
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o')  # 위와 동일한 문장, 더 명시적
````

````python
data = np.random.randn(30).cumsum()
plt.plot(data, 'y--', label='Flow')  # 여기까진 점선그래프 생성
plt.plot(data, 'b--', drawstyle='steps-post', label='steps')
plt.legend(loc='best')
````

````python
#리스트 작성
lang.append('python')
lang.append('java')
lang.append('c')
lang.append('R')
lang
````

````python
#리스트에서 인덱스로 꺼내기
#-1은 마지막 인덱스 꺼내기
lang[0], lang[1], lang[-1]
````

````python
#반복문으로 리스트 원소 하나씩 출력
for i in lang:
    print(i)
````

````python
#들여쓰기(인덴트) 꼭 주의
for i in lang:
    if i == 'python':
        print('like')
    else:
        print('nope')
````

````python
#특정 회수만큼 반복문 실행
for i in range(3):
    print(lang[i])
````

````python
#랭 리스트의 길이를 변수로 받아서 그만큼 반복
count = len(lang)
for i in range(count):
    print(lang[i])
````

````python
#짝수 일때 파이썬을 홀수일때 자바를 출력
for i in range(1,10):
    if i % 2 == 0:
        print('파이썬')
    else:
        print('자바')
````

````python
# enumerate로 인덱스번호와 값을 같이 출력
for i, val in enumerate(lang):
    print(i, val)
````

### pandas

````python
import pandas as pd
````

````python
df = pd.DataFrame(
{'a' : [4,5,6, 4],
 'b' : [7,8,9, 9],
 'c' : [10,11,12, 11]},
index = [1,2,3,4])
# 데이터 프레임 출력(2차원)
df
````

````python
# 시리즈 데이터 출력(1차원)
df['a']
````

````python
# 시리즈를 데이터 프레임으로 출력
df[['a']]
````

````python
#조건 논리 확인
df['a'] > 4
````

````python
#조건에 만족하는 데이터 프레임만 가져옴
df[df['a'] > 4]
````

````python
#2개 이상의 컬럼을 가져올때는 시리즈 출력불가, 데이터 프레임으로 가져오기
df[['a', 'b']]
````

````python
#summarize data
df['a'].value_counts()
````

````python
len(df)
````

````python
#컬럼 삭제
df = df.drop(['c'], axis=1)
df
````

````python
#a컬럼 그룹바이해서 b컬럼 값으로 평균을 구하겠다
df.groupby(['a'])['b'].mean()
````

````python
# 각종 연산 정보
df.groupby(['a'])['b'].agg(['mean', 'sum', 'count'])
````

````python
# 요약 정보 한번에 보기
df.groupby(['a'])['b'].describe()
````

````python
#a컬럼 기준으로 
pd.pivot_table(df, index='a')
````
